```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
from datetime import datetime
import re
```

```python
url = 'http://stage48.net/wiki/index.php/BNK48_Members'
response = requests.get(url,verify = False)
htmlText = response.content.decode('UTF-8')
soup = bs(htmlText,'html.parser')

table = soup.find_all('table')
memberlist = table[2].find_all('tr')
test1 = memberlist[1].find_all('td')
test1
#test2 = memberlist[1].find_all('th')
#test2[0].find_all('span')
```

```python
count2 = 0
member = []
for temptable in table:
    if count2 == 0:
        count2+=1
        continue
    count = 0
    regex = re.compile(r'([ก-๙]+)', re.I)
    for i  in temptable.find_all('tr'):
        d = {}
        if count ==0:
            count+=1
            continue
        #print(i)
        data = i.find_all('td')
        data2 = i.find('span')
        d['team'] = data2.text.strip()
        d['name'] = data[2].text.strip()
        d['name'] = regex.findall(d['name'])
        d['name'] = d['name'][0]+" "+d['name'][1]
        d['nickname'] = data[3].text.strip()
        d['nickname' ] = regex.findall(d['nickname'])[0]
        d['birth'] = data[4].text.strip().replace('Junuary', 'January')
        d['birth'] = datetime.strptime(d['birth'], '%B %d, %Y')
        d['birth'] = d['birth'].date().isoformat()
        d['age'] = int(data[7].text.strip())
        d['gen'] = data[10].text.strip()
        member.append(d)
        #print(member)
        #print("------------------------------------------")
member
```

```python
df = pd.DataFrame(member)
df = df[['team','nickname','name','birth','gen','age']]
df
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
sns.set(style = "darkgrid")
ax = sns.countplot(x='age',data = df)
mean = np.average(df['age'])
med  = np.median(df['age'])
mode = stats.mode(df['age'])
print(mean)
print(med)
print(mode)
```
