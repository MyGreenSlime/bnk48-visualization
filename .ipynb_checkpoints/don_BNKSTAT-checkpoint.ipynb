```python
import numpy as np
import re
member = np.unique(df['name'])
print(member)
fil = ['bnk48official.aom','bnk48official.bamboo','bnk48official.cake','bnk48official.deenee','bnk48official.faii','bnk48official.fifa','bnk48official.fond','bnk48official.gygee','bnk48official.jan','bnk48official.june','bnk48official.khamin','bnk48official.kheng','bnk48official.maira','bnk48official.mewnich','bnk48official.minmin','bnk48official.myyu','bnk48official.natherine','bnk48official.new','bnk48official.niky','bnk48official.nine','bnk48official.oom','bnk48official.panda','bnk48official.phukkhom','bnk48official.ratah','bnk48official.stang','bnk48official.view','bnk48official.wee','pakwan.bnk48official']
print(len(fil))
memberx = list(member)
member = []
for i in memberx:
    if i not in fil:
        member.append(i)
member
```

```python
memberdata = {}
for i in member:
    memberdata[i] = list()
memberdata
for j,i in df.iterrows():
    d ={}
    fuck = i['datetime (utc timezone)'].split(' ')
    #print(fuck)
    if(re.match('13:00:[0-9]',fuck[1]) and (fuck[0] != '2018-04-28' and fuck[0] != '2018-04-27'and fuck[0] !='2018-05-29')):
        d['date'] = i['datetime (utc timezone)']
        d['like'] = i['fan_count']
        memberdata[i['name']].append(d)
memberdata
```

```python
import matplotlib.pyplot as plt
for mem in member:
    x =[]
    y =[]
    for i in memberdata[mem]:
        x.insert(0,i['date'])
        y.insert(0,i['like'])
    plt.plot(x,y)
    plt.show()
    print(mem)
    arange = y[-1]-y[0]
    print("RANGE : ",arange)
    growthperday = arange/30
    print("GROWTH PER DAY"," : ",growthperday)
    percent = ((y[-1]/y[0])*100)-100
    print("GROWTH PERCENT",percent)
```
