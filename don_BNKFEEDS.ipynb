```python
url = 'https://graph.facebook.com/v3.0/'
filed = "/posts/?fields=message,id,type,permalink_url,comments.limit(0).summary(true),shares,reactions.limit(0).summary(true)&access_token="+token
member = {}
for i in namemember:
    member[i] = []
    realurl = url+i+filed
    re = requests.get(realurl)
    feed =  re.json()
    #print(feed)
    print(i)
    while 'next'or'previous' in feed['paging']:
        for f in feed['data']:
            data = {}
            print(f['id'])
            try:
                data['id'] = f['id']
            except:
                data['id'] = None
            try:
                data['message'] = f['message']
            except:
                data['message'] = None
            data['type'] = f['type']
            try:
                data['commentcount'] = int(f['comments']['summary']['total_count'])
            except:
                data['commentcount'] = 0
            
            try:
                data['reaction'] = int(f['reactions']['summary']['total_count'])
            except:
                data['reaction'] = 0
            try:
                data['share'] = int(f['shares']['count'])
            except:
                data['share'] = 0 
            data['link'] = f['permalink_url']
            member[i].append(data)
        try:
            realurl = feed['paging']['next']
            re = requests.get(realurl)
            feed =  re.json()
        except:
            break
    

```

```python
statall = []
for i  in namemember:
    statmem ={}
    statmem['name'] = i.split(".")[1]
    sumreaction = 0
    sumshare = 0
    sumcomment = 0
    maxcomment = 0
    maxshare = 0
    maxreaction = 0
    count = 0
    idremax = ''
    idcomax =''
    idshamax = ''
    typepost = {}
    typepost['photo'] = 0
    typepost['link'] = 0
    typepost['status'] = 0
    typepost['video'] = 0
    for data in member[i]:
        try:
            typepost[data['type']]+=1
        except:
            pass
        sumreaction+=data['reaction']
        sumshare+=data['share']
        sumcomment+=data['commentcount']
        if(maxreaction <= data['reaction']):
            maxreaction  = int(data['reaction'])
            idremax = data['id']
        if(maxcomment <= data['commentcount']):
            maxcomment = int(data['commentcount'])
            idcomax = data['id']
        if(maxshare <= int(data['share'])):
            maxshare =int(data['share'])
            idshamax = data['id']
        count+=1
        
    statmem['totalpost'] = count
    statmem['totalreaction'] = sumreaction
    statmem['totalshare'] = sumshare
    statmem['totalcomment'] = sumcomment
    statmem['averreaction'] = sumreaction/count
    statmem['avercomment'] = sumcomment/count
    statmem['avershare'] = sumshare/count
    statmem['maxreaction'] = maxreaction
    statmem['idremax'] = idremax
    statmem['maxcomment'] =maxcomment
    statmem['idcomax'] = idcomax
    statmem['maxshare'] = maxshare
    statmem['idshamax'] = idshamax
    for ty in typepost:
        statmem[ty] = int(typepost[ty])
    statall.append(statmem)
    
        
```
